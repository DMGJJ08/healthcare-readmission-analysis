```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('diabetic_data.csv')
print(df.shape)
df.head()
```

```python
# Replace '?' with NaN and drop missing
df = df.replace('?', pd.NA)
df = df.dropna(subset=['race', 'gender', 'age'])

# Remove unnecessary columns
df = df.drop(columns=['weight', 'payer_code', 'medical_specialty'])

# Simplify readmitted column
df['readmitted'] = df['readmitted'].map({'<30': 'Readmitted <30 days', '>30': 'Readmitted >30 days', 'NO': 'Not Readmitted'})

print(df.shape)
print(df['readmitted'].value_counts())
```

```python
# Chart 1 - Readmission rates by age group
plt.figure(figsize=(12, 5))
age_readmit = df.groupby('age')['readmitted'].value_counts(normalize=True).unstack()
age_readmit['Readmitted <30 days'].sort_index().plot(kind='bar', color='crimson')
plt.title('Readmission Rate Within 30 Days by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Readmission Rate')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('readmission_by_age.png')
plt.show()
```

```python
# Chart 2 - Readmission by number of medications
plt.figure(figsize=(12, 5))
sns.boxplot(data=df, x='readmitted', y='num_medications', palette=['green', 'orange', 'crimson'])
plt.title('Number of Medications vs Readmission Status')
plt.xlabel('Readmission Status')
plt.ylabel('Number of Medications')
plt.tight_layout()
plt.savefig('medications_vs_readmission.png')
plt.show()
```

```python
# Chart 3 - Time in hospital vs readmission
plt.figure(figsize=(12, 5))
sns.countplot(data=df, x='time_in_hospital', hue='readmitted', 
              palette=['green', 'orange', 'crimson'])
plt.title('Length of Hospital Stay vs Readmission Status')
plt.xlabel('Days in Hospital')
plt.ylabel('Number of Patients')
plt.legend(title='Readmission')
plt.tight_layout()
plt.savefig('stay_vs_readmission.png')
plt.show()
```

```python
# Chart 4 - Top diagnoses linked to readmission
top_diag = df[df['readmitted'] == 'Readmitted <30 days']['diag_1'].value_counts().head(10)

plt.figure(figsize=(12, 5))
top_diag.plot(kind='bar', color='crimson')
plt.title('Top 10 Primary Diagnoses for 30-Day Readmissions')
plt.xlabel('Diagnosis Code')
plt.ylabel('Number of Readmissions')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('diagnoses_readmission.png')
plt.show()
```
